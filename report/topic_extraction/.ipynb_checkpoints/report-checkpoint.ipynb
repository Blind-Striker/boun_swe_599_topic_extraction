- `numpy`: A library for the Python programming language, adding support for large, multi-dimensional arrays and matrices, along with a large collection of high-level mathematical functions to operate on these arrays.

- `collections`: This module implements specialized container datatypes providing alternatives to Pythonâ€™s general purpose built-in containers.
- `json`: This module is used to work with JSON data.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
import json
```

## Data normalization
In this cell, the data is being read from a JSON file and then normalized. Normalization is the process of organizing the data in a way that aids the extraction of meaningful insights. In this case, the 'Annotations' field in the data is being normalized, which will allow for a more efficient analysis of the different annotations of each tweet.


```python
with open('E:\\buon\\swe_599\\boun_swe_599_topic_extraction\\data\\sentiment_20230607181851783_in_json.json') as f:
    data = json.load(f)

# Normalize the Annotations field
df = pd.json_normalize(data, record_path='Annotations', meta=['date', 'favorite', 'replies', 'retweets', 'unix_timestamp', 'url'])

df.head()
```

## Rho vs link_probability
A scatterplot of 'rho' and 'link_probability' values is plotted in this cell to find the most relevant values to filter the data with. 'rho' and 'link_probability' are measures of confidence that a given annotation is correct, so by plotting them, we can understand the distribution of these confidence scores in the dataset and choose appropriate thresholds for filtering.


```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='rho', y='link_probability', data=df)
plt.title('rho vs link_probability')
plt.show()
```

## Most frequent topics
After filtering the data based on 'rho' and 'link_probability', the most frequent Wikidata topics are listed. This is done by:

1. Filtering the dataframe for rows that have a 'link_probability' and 'rho' greater than the determined threshold values. This leaves us with the annotations that the model is most confident about.
2. The 'WikidataTopics' field is exploded to create separate rows for each topic linked to a single annotation.
3. The 'Value' field is extracted from each dictionary in 'WikidataTopics' to give us the name of the topic.
4. A value count is performed on the 'Topic' field to find the most frequent topics.


```python
# 1. Filter out rows based on `link_probability` and `rho`
filtered_df = df[(df['link_probability'] >= 0.1) & (df['rho'] >= 0.3)]
# 2. Analyse topics
# Explode the list across rows
topics_df = filtered_df.explode('WikidataTopics')

# Extract the 'Value' field from each dictionary in 'WikidataTopics'
topics_df['Topic'] = topics_df['WikidataTopics'].apply(lambda x: x['Value'])

# Now you can do a value_counts to find the most frequent topics
print(topics_df['Topic'].value_counts().head(10))
```

## Outlier removal
The 'large language model' and 'generative pre-trained transformer' topics are found to be exceptionally frequent. This is because the tweets in the data were related to the ChatGPT model, so these topics are expected to be common. Hence, they are treated as outliers and removed from the data to prevent them from overshadowing other potentially interesting topics.


```python
topics_to_exclude = ['large language model', 'generative pre-trained transformer']
topics_df = topics_df[~topics_df['Topic'].isin(topics_to_exclude)]
print(topics_df['Topic'].value_counts().head(20))
```

## Top 50 topics visualization
After excluding the outliers, a bar plot of the top 50 topics is created. This visualization helps us understand which topics, apart from the excluded ones, are the most frequently mentioned in the data.

```python
top_50_topics = topics_df['Topic'].value_counts().head(50)
top_50_topics.plot(kind='bar', figsize=(15,7))

plt.title('Top 50 Topics')
plt.xlabel('Topic')
plt.ylabel('Frequency')

plt.show()
```

## Topic categorization and visualization
Given the wide range of topics obtained from the data, it is helpful to group them into broader categories to gain better insights. Therefore, a manual categorization is created to group similar topics under a common category. After assigning each of the top 50 topics to a category, a bar plot is created to visualize the total count of topics under each category. This helps us understand the most frequently discussed categories in the data.


```python
# Define topic categories
topic_categories = {
    "AI and Machine Learning": ["artificial intelligence", "machine learning", "deep learning",
                                "chatterbot", "a.i. artificial intelligence", "machine learning method"],
    "Computer Science": ["computer science", "python (programming language)", "software agent",
                         "computing platform", "computer interface", "communication protocol",
                         "application programming interface", "Unix-like operating system"],
    "Financial and Legal": ["legal tender", "S&P 500", "investment", "bitcoin", "payment system",
                            "Dow Jones Industrial Average", "financial sector", "financial technology"],
    "Cryptocurrency": ["crypto asset", "cryptocurrency", "digital currency"],
    "Social Media and Technology": ["openai", "Big Tech (web)", "twitter", "virtual world", "metaverse",
                                    "microsoft", "technology", "youtube", "whatsapp"],
    "Data Science": ["data", "big data", "data science"],
    "Robotics and Automation": ["robotics", "control engineering, mechatronics and robotics", "mechatronics"],
    "Economics and Business": ["emerging technology", "Nasdaq-100", "industrial sector"],
    "Science and Mathematics": ["cognitive science", "mathematics", "science"],
    "Web Development": ["javascript"],
    "Health and Medicine": ["medicine"]
}

# Create a new dataframe from the top 50 topics
top_50_topics_df = pd.DataFrame(top_50_topics).reset_index()
top_50_topics_df.columns = ['Topic', 'Count']

# Create a new column for the topic category
def assign_topic_category(topic):
    for category, topics in topic_categories.items():
        if topic in topics:
           return category
    return "Other"

top_50_topics_df['Topic Category'] = top_50_topics_df['Topic'].apply(assign_topic_category)

# Count the occurrences of each category
topic_category_counts = top_50_topics_df.groupby('Topic Category')['Count'].sum()

# Sort the topic categories in ascending order based on total count
topic_category_counts = topic_category_counts.sort_values(ascending=True)

# Plot
fig, ax = plt.subplots(figsize=(10, 7))
bar_plot = topic_category_counts.plot(kind='barh', ax=ax)

# Add tooltips
for rect in bar_plot.containers[0]:
    width = rect.get_width()
    category = topic_category_counts.index[bar_plot.containers[0].index(rect)]
    ax.annotate(f'{width}', xy=(width, rect.get_y() + rect.get_height() / 2),
                xytext=(3, -6),  # 3 points horizontal offset, -6 points vertical offset
                textcoords="offset points",
                ha='left', va='center')

plt.title('Total Number of Topics by Category')
plt.xlabel('Total Count')
plt.ylabel('Topic Category')
plt.show()
```

## Data normalization
In this cell, the data is being read from a JSON file and then normalized. Normalization is the process of organizing the data in a way that aids the extraction of meaningful insights. In this case, the 'Annotations' field in the data is being normalized, which will allow for a more efficient analysis of the different annotations of each tweet.


```python
## Normalize the Annotations field
df_sentiments = pd.json_normalize(data, record_path='Sentiments', meta=['date', 'favorite', 'replies', 'retweets', 'unix_timestamp', 'url'])

df_sentiments.head()
```

## Sentiment Analysis
In this cell, we perform sentiment analysis on our dataset. First, we normalize the 'Sentiments' column of our dataset to create a separate row for each sentiment label ('Positive', 'Negative', and 'Neutral') and its corresponding score. Each tweet has an associated sentiment score which represents the level of positivity, negativity, or neutrality of the text content in the tweet. This normalization helps us analyze each sentiment separately in the following steps.


```python
# First, convert 'score' column to float
df_sentiments['score'] = df_sentiments['score'].astype(float)

# Calculate the mean sentiment score for each sentiment type
mean_scores = df_sentiments.groupby('label')['score'].mean()

# Now let's plot this information
mean_scores.plot(kind='bar', figsize=(7,5))
plt.title('Mean Sentiment Scores')
plt.xlabel('Sentiment')
plt.ylabel('Score')
plt.show()
```

## Sentiment Analysis Over Time
This cell generates a visualization of sentiment scores over time. By doing this, we can track how sentiments related to our topic of interest (e.g cryptocurrency) have changed over time. The x-axis represents time (from the last quarter of 2022 to today), while the y-axis shows the average sentiment scores. Each sentiment (Positive, Negative, Neutral) is represented by a separate line in the plot.


```python
# Convert 'unix_timestamp' column to datetime format, in milliseconds
df_sentiments['date'] = pd.to_datetime(df_sentiments['unix_timestamp'], unit='ms')

# Filter the dataframe for entries from last quarter of 2022 to today
df_sentiments = df_sentiments[(df_sentiments['date'] >= '2022-11-20') & (df_sentiments['date'] <= pd.Timestamp.today())]

# Group by date and sentiment label, then calculate mean score for each group
sentiment_over_time = df_sentiments.groupby([df_sentiments['date'].dt.date, 'label'])['score'].mean().reset_index()

# Pivot the data so that we have one column for each sentiment label
pivot_table = sentiment_over_time.pivot(index='date', columns='label', values='score')

# Plot the data
pivot_table.plot(kind='line', figsize=(10,5))
plt.title('Sentiment Score Trends Over Time')
plt.xlabel('Date')
plt.ylabel('Average Sentiment Score')
plt.show()
```

## Prepare Data for Topic-Sentiment Correlation
This cell is used to prepare the data for the correlation between topics and sentiments. Here, we merge our 'topics' dataframe with our 'sentiments' dataframe on the 'url' column. We then group the merged dataframe by 'Topic' and 'label', and calculate the mean score for each group. Afterward, we pivot the dataframe to have one column for each sentiment label. We sort the dataframe in descending order based on frequency and select the top 50 topics.


```python
# Merge the two dataframes on the 'url' column
merged_df = pd.merge(topics_df, df_sentiments, on='url')

# Filter to include only the top 50 topics
filtered_df = merged_df[merged_df['Topic'].isin(top_50_topics.index)]

# Calculate the frequency of each topic
topic_counts = filtered_df['Topic'].value_counts()

# Group by Topic and label, then calculate mean score for each group
topic_sentiment = filtered_df.groupby(['Topic', 'label'])['score'].mean().reset_index()

# Pivot the data so that we have one column for each sentiment label
pivot_table = topic_sentiment.pivot(index='Topic', columns='label', values='score')

# Create a 'total' column with the mean sentiment score
pivot_table['total'] = pivot_table.mean(axis=1)

# Reorder the pivot_table index according to the topic frequency
pivot_table = pivot_table.reindex(topic_counts.index)
```

## Correlation Between Topics and Sentiments
This cell produces a visualization showing the correlation between topics and sentiments. It displays the sentiment scores (Positive, Negative, Neutral) for the top 50 categories (based on frequency) instead of individual topics. The 'total' column representing the mean sentiment score is dropped for the visualization. The bar plot helps us understand the average sentiment score associated with each topic, thus allowing us to identify which categories are generally perceived more positively, neutrally, or negatively.


```python
# Display the top 50 topics by total mean sentiment score
pivot_table.head(50).drop('total', axis=1).plot(kind='bar', stacked=True, figsize=(20,10))
plt.title('Average Sentiment Scores for Top 50 Topics')
plt.xlabel('Topic')
plt.ylabel('Average Sentiment Score')
plt.show()
```

## Correlation Between Categories and Sentiments

This cell generates a visualization that represents the correlation between the categories and sentiments. Instead of using individual topics, we classify the topics into broader categories such as 'AI and Machine Learning', 'Computer Science', 'Financial and Legal', etc. Categorization of topics allows us to view the data more broadly and comprehensively.

```python
# Flatten the dictionary
categories_dict = {topic: category for category, topics in topic_categories.items() for topic in topics}

# Map 'Topic' to 'Category' using categories_dict
merged_df['Category'] = merged_df['Topic'].map(categories_dict)

# Group by Category and label, then calculate mean score for each group
category_sentiment = merged_df.groupby(['Category', 'label'])['score'].mean().reset_index()

# Pivot the data so that we have one column for each sentiment label
pivot_table = category_sentiment.pivot(index='Category', columns='label', values='score')

# Calculate the frequency of each category
category_counts = merged_df['Category'].value_counts()

# Create a 'total' column with the mean sentiment score
pivot_table['total'] = pivot_table.mean(axis=1)

# Reorder the pivot_table index according to the category frequency
pivot_table = pivot_table.reindex(category_counts.index)

# Display the top 50 categories by total mean sentiment score
pivot_table.head(50).drop('total', axis=1).plot(kind='bar', stacked=True, figsize=(20,10))
plt.title('Average Sentiment Scores for Categories')
plt.xlabel('Category')
plt.ylabel('Average Sentiment Score')
plt.show()
```
